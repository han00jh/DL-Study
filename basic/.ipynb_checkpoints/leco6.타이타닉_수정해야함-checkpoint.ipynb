* 80% 문제지 답안지

````python
# 1. 빈칸(NaN) 강제로 다 채우기
X80 = X80.fillna(0)
X20 = X20.fillna(0)

# 2. 이제 다시 학습시키기 (오타 없으니 그대로 실행!)
lr.fit(X80, y80)
dt.fit(X80, y80)
````

````python
lr.fit(X80, y80)
dt.fit(X80, y80)
````
