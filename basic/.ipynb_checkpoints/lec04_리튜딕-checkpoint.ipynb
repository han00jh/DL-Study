## 넘파이 배열

```python
import numpy as np
myarr = np.array( [1,2,3] )
print( myarr )
```

```python
print( myarr[0] )
```

## 리스트 []

```python
mylist = [1,2,3]
print(mylist)
```

### **조회**

```python
mylist[0]
```

### **입력**

```python
mylist.append(4)
```

```python
mylist
```

### 수정

```python
mylist.insert(0,99)
```

```python
mylist
```

### 삭제

```python
del mylist[2]
```

```python
mylist
```

## 딕셔너리 {}

### 조회

```python
mydic = {"name":"홍길동" ,  "age":20}
print( mydic )
```

```python
mydic["name"]
```

### 입력
* 없는 키값 넣으면 입력

```python
mydic["tel"] = "010"
mydic
```

### 수정
* 있는 키값 넣으면 수정

```python
mydic["name"] = "김길동"
mydic
```

### 삭제

```python
del mydic["name"]
mydic
```

```python
for t in mydic.items():
    print(t[0],t[1])
```

```python
mydata = [[1,2,3], [10,20,30]]
for dan in mydata:
   # print(dan)
    for gob in dan:
        print(gob)
```

```python
mydata = [
              {"name":"kim" ,  "age":10}, 
              {"name":"hong",  "age":20}
         ]
for dic in mydata:
    #print(dic.values())
    print(dic["name"], dic["age"])
```

```python
mydata[1]["age"]
```

```python
mydata = [
              {"name":"kim" ,  "age":10}, 
              {"name":"hong",  "age":20}
         ]
for dic in mydata:
    print(dic.values())
    for v in dic.values():
        print(v)
        
```

```python
mydata= { "name" : ["kim","hong"]  ,
          "age" : [10, 20]
        }

for dic in mydata.items():
    print(dic[0],dic[1])
```

```python
mydata["name"][1]
```

```python
mydata= { "name" : ["kim","hong"]  ,
          "age" : [10, 20]
        }
for key in mydata.keys():
    print(key, mydata[key])
```

```python
for val in mydata.values():
    print(val)
```

```python
fruit_box = {
    "apple": 1000,
    "banana": 2500,
    "cherry": 5000
}

for item in fruit_box.items():
    if item[0] == "apple":
        print(f"{item[0]}가격은{item[1]}원입니다.")

```

```python
mydata= { "name" : ["kim","hong"]  ,
          "age" : [10, 20]
        }
for vs in mydata.values():
        print(vs)
```

```python
mydata= { "name" : ["kim","hong"]  ,
          "age" : [10, 20]
        }
for vs in mydata.values():
    for v in vs:
        print(v)
```

```python
mydata = [  
            [  {"name":"kim" ,  "age":10}, 
               {"name":"hong",  "age":20}  
            ]
         ]
for key in mydata:
    for v in key:
        print(v)
```

```python
mydata = [  
            [  {"name":"kim" ,  "age":10}, 
               {"name":"hong",  "age":20}  
            ]
         ]
for sub_list in mydata:
    for dic in sub_list:
        for val in dic.values():
            print(val)
```

```python
mydata = [  
            [  {"name":"kim" ,  "age":10}, 
               {"name":"hong",  "age":20}  
            ]
         ]
for sub_list in mydata:
    for dic in sub_list:
        print(dic["name"],dic["age"])
```
