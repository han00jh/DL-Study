## <b> 컬럼1 집계N

````python
emp.groupby('DEPTNO')['SAL'].agg( ['mean','max','sum'] )
````

## <b> 컬럼N 집계N
<font color=red><b>emp.groupby('DEPTNO')[['SAL','EMPNO']].agg(['min','max','sum'])

````python
emp.groupby('DEPTNO') [['SAL','EMPNO']] .agg( ['min' ,'max', 'sum'] )
````

## <b> 컬럼N 별 다른 집계
<font color=red><b>emp.groupby('DEPTNO').agg({'SAL' : ['min','max'], 'EMPNO': ['count']})

* <b>select deptno, max(sal), min(sal), count(empno) from emp group by deptno;

````python
emp.groupby('DEPTNO').agg({'SAL' : ['min','max'], 'EMPNO': ['count']})
````

# 데이터 프레임 합치기

#### <font color=blue> 컬럼이름 동일, 인덱스값 동일

````python
df1 = pd.DataFrame([ [1,10]
                   , [2,20]
                  ] , columns=["eno","sal"],index=[0,1])
df2 = pd.DataFrame( [['a','aa'] 
                   , ['b','bb']
                  ]  , columns=["eno","sal"],index=[0,1])
````

````python
df1
````

````python
df2
````

````python
df3 = pd.concat([df1,df2], axis = 0) #행(줄)방향합치기 = 세로
df3
````
