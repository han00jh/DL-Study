<pre><b><font color=red size=4>re.함수('패턴' , '대상문장')

```python
res = re.match('[a-z]', "ABC")
res
```

```python
res = re.match('[a-z]', "abc") #문자열의 처음만보고 판단
res 
```

```python
res = re.search('[a-z]', "1abc") #문자열의 문장전체를 보고 찾으면 매치
res
```

```python
res = re.findall('[a-z]', "1abc") #문자열의 문장전체를 보고 찾았어도 계속 찾음 --[]
res
```

```python
resobj = re.finditer('[a-z]', "1abc") #문자열의 문장전체를 보고 찾았어도 계속 찾음--객체
resobj
```

```python
for r in resobj: #일회용함수라 여러번출력못함
    print(r) # 값만 빼고싶으면 r.group() 또는 r[0]
```

## 이메일찾기
* abc@aa.com
* lee.kim@korea.com

[\w\.-]+@[\w\.-]+\.\w+

```python
p1 = re.compile("[\w\.-]+@[\w\.-]+\.\w+")
res = p1.match('abc@aa.com')
print(res)

p = re.compile('[a-z]*[.a-z]*@[A-Za-z]+\.[A-Za-z]{3,}')
res = p.match('lee.kim@korea.com')
print(res)
```

## 전화번호
* 010-123-1234
* 010-4561-4545
* 02-123-7895

```python
p = re.compile('\d{2,3}-\d{3,4}-\d{4}')
res = p.match('010-1234-1234')
print(res)

p = re.compile(r"^0(2|1[0-9])\-[0-9]{3,4}\-[0-9]{4}$")
res = p.match("010-4561-4545")
res
```

## 파일 확장자 찾기
* aa.txt aa.jpg a100.txt   -->  txt  jpg  txt

```python
mylist = ["aa.txt","aa.jpeg","a100.txt"]
for fname in mylist:
    flist = fname.split(".")
    print(flist[1])
```

* <font color=red> () : 캡쳐

```python
p= re.compile("\w+\.(txt)")
res = p.findall("aa.txt aa.job a100.txt")
res
```

* 02-123-1234  032-5151-1234 032-1234-5454 031-1234-5454

```python
p= re.compile('(\d+)-\d+-\d+')
res = p.findall("02-123-1234 032-5151-1234 032-1234-5454 031-1234-5454")
res
```

```python
p = re.compile('\.([a-z]+)')
res = p.findall('aa.txt aa.jpg a100.txt')
print(res)

# 확장자가 txt인 파일 출력 - findall()

p1   = re.compile(r"\w+\.txt")
p1_1 = re.compile(r"\w+\.txt$")
p2   = re.compile("\w+\.txt")
p2_1 = re.compile("\w+\.txt$")

res_1   = p1  .findall("aa.txt aa.job a100.txt")
res_1_1 = p1_1.findall("aa.txt aa.job a100.txt")
res_2   = p2  .findall("aa.txt aa.job a100.txt")
res_2_1 = p2_1.findall("aa.txt aa.job a100.txt")

res_1
res_1_1
res_2
res_2_1
```

## 파일 확장자가 txt인 파일 찾기
* aa.txt aa.jpb a100.txt   -->  aa.txt   a100.txt

```python
p = re.compile('\w+\.txt')
res = p.findall('aa.txt aa.jpg a100.txt')
print(res)
```

## 호칭 찾기

```python
name1 = "Braund, Mr. Owen Harris"
name2 = "Cumings, Mrs. John Bradley (Florence Briggs Thayer)"
name3 = "Heikkinen, Miss. Laina"
```

```python
#p = re.compile('([A-Za-z]+)\.')
p = re.compile(',\s([A-Za-z]+\.)\s')

res = p.findall(name1)
print(res)
res = p.findall(name2)
print(res)
res = p.findall(name3)
print(res)
```

## 한글찾기

```python
p = re.compile("[가-힣]+")
res = p.findall("The name is Jack, kim, 홍길동, king, 김길동")
res
```

```python
p = re.compile("[가-힣ㄱ-ㅎㅏ-ㅣ]+")
res = p.findall("The name is Jack, kim, 홍길동, king, 김길동 ㅋㅋㅋ ㅠㅠㅠ ㅎㅎㅎ")
res
```
