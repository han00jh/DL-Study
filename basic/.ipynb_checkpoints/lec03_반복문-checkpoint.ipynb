## 중첩 for

```python
for dan in range(2,6):
    print(f"{dan}단")
    for gob in range(1,10):
        print(f"{dan}*{gob}={dan*gob}" ,end ="\t")
    print()
```

## 1~10까지 짝수의 합

```python
tot =0
for num in range(2,11,2):
    tot+=num
print(tot)
    
```

```python
tot = 0
for num in range(1,11):
    if(num%2 ==0):
        tot += num
print(tot)
```

```python
tot = 0
oto = 0
for num in range(1,11):
    if(num%2 ==0):
        tot += num
    else:
        oto += num
print(f"짝총합 :{tot}")
print(f"짝총합 :{oto}")
```
