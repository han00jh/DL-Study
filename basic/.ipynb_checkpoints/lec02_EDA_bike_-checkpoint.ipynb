### <b> ★★ 상관분석 ★★
<font size=3> <b>📊공분산과 상관계수 비교 요약

| 구분 | 공분산 (Covariance) | 상관계수 (Correlation Coefficient) |
| :--- | :--- | :--- |
| **기호** | $Cov(X, Y)$ | $\rho_{XY}$ |
| **핵심 정의** | 두 변수가 함께 변하는 정도 | 공분산을 표준화하여 관계의 방향과 강도를 나타낸 값 |
| **수치 범위** | 제한 없음 (데이터 단위에 의존) | $-1 \le \rho_{XY} \le 1$ |
| **양의 관계 ($>0$)** | $X$ 증가 시 $Y$ 증가 | 1에 가까울수록 강한 양의 선형 관계 |
| **음의 관계 ($<0$)** | $X$ 증가 시 $Y$ 감소 | -1에 가까울수록 강한 음의 선형 관계 |
| **관계 없음 ($\approx 0$)** | 선형 관계 없음 | 선형 관계 없음 |
| **공식** | $E[(X - \mu_X)(Y - \mu_y)]$ | $\frac{Cov(X, Y)}{\sigma_X\sigma_Y}$ |
| **주요 차이점** | 방향성만 제시함 | 방향성과 **강도**를 모두 제시함 |

> **Note:** 자기 자신과의 공분산 $Cov(X, X)$은 해당 변수의 분산 $Var(X)$과 같습니다.

<br>
<font size=3> <b>📊다중공선성 (Multicollinearity) 요약</b>

| 구분 | 내용 설명 |
| :--- | :--- |
| **정의** | 독립 변수(Feature)들끼리 서로 강한 상관관계를 가지는 현상 |
| **비유** | 두 명의 사수가 같은 과녁을 쏘는데, 누구 실력인지 구분하기 힘든 상태 |
| **발생 원인** | `temp`와 `atemp`처럼 본질적으로 같은 정보를 담은 변수를 동시에 넣을 때 |
| **주요 문제** | 1. 모델의 예측 성능 저하 및 불안정성 <br> 2. 어떤 변수가 정답에 기여하는지 해석 불가능 |
| **판단 기준** | 1. 상관계수: 히트맵에서 수치가 **0.9 이상**인 경우 <br> 2. VIF 지수: **10 이상**이면 심각한 공선성으로 판단 |
| **해결 방법** | 1. **변수 제거**: 상관관계가 높은 변수 중 하나를 삭제 <br> 2. **변수 통합**: 두 변수를 평균 내거나 차이값으로 변환 |

#### 방법2) np.where(조건절, 참, 거짓)
* decode(deptno,   10,1   ,20,2   ,3)
* decode(deptno,   10,       1   ,99)

````python
train['day_type'] =0
np.where((train['holiday']==0) & (train['workingday']==1), 0, train['day_type'])
np.where((train['holiday']==0) & (train['workingday']==0), 1, train['day_type'])
np.where((train['holiday']==1) & (train['workingday']==0), 2, train['day_type'])
````

#### 방법3) lambda

````python
traincp = train.copy()
````

````python
traincp['day_type3'] =0
def myfunc(traincp):
    day_type =0
    if (traincp['holiday']==0) and (traincp['workingday']==1):  #평일
        day_type=0
    elif  (traincp['holiday']==0) and (traincp['workingday']==0): #주말
        day_type=1
    elif   (traincp['holiday']==1) and (traincp['workingday']==0): #공휴일
        day_type=2
    return day_type

traincp['day_type3'] = traincp.apply(lambda df : myfunc(df), axis =1)
````

## <font color =red><b>★★★★★★★정규화 ★★★★★★★

<pre><font color=red><b>
* Regularization
    * 모델의 복잡도를 제어 -> 과적합(Overfitting)을 방지 ->모델의 일반화
    * 일부를 희생 -> 예측 성능을 높이는데 표현
 * Normalization == 스케일링(scaling)
    *  데이터의 스케일을 조정(0~1) -> 피쳐의 영향력을 줄임 -> 모델학습 안정화
    * 0 ~ 1 사이의 값으로 변환

### 스케일링

````python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
````

#### <b><font color=red> MinMaxScaler()
* 최대(1) ~ 최소(0)범위로 축소

````python
col_list =  ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'count']

mm_scaler = MinMaxScaler()
mm_scaler.fit(train[col_list]) #데이터넣을때 2차
scaler_res = mm_scaler.transform(train[col_list])
scaler_res
````

#### <pre><b><font size=5 color =red> ★★★★★★★★★StandardScaler()★★★★★★★★★
* 평균 0으로 범위 축소

<b><font size=4 color=red>Z-값(Z-score, 표준점수) <br>
   값이 평균으로부터 표준편차의 몇 배만큼 떨어져 있는지를 나타내는 수치<br>
   평균 0, 표준편차 1인 표준정규분포 <br>
    $z=\frac{x-\mu }{\sigma }$ = 오차(실값-평균)를 표준편차로 표준화

#### <b><font color=red> RobustScaler()
* 중앙값 기준으로 IQR 1사분위(0.25) ~ 3사분위(0.75) 범위로 축소 

````python
col_list =  ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'count']
scaler_list = [StandardScaler(), MinMaxScaler(), RobustScaler()]
for scaler in scaler_list:
    scaler.fit(train[col_list])
    res = scaler.transform(train[col_list])
    print(scaler.__class__.__name__, res[:2])
````

## 1차 우선 점수만 보기

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split

x= train.drop(['count','casual','registered'], axis=1)
y= train['count']
x80, x20, y80, y20 = train_test_split(x, y, test_size=0.2, random_state=0)
print(x80.shape, x20.shape, y80.shape, y20.shape)
````

````python
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr.fit(x80, y80)
pred = lr.predict(x20)
score = mean_squared_error(y20, pred)
print( np.sqrt(score) )
print(y20[:10])
print(pred[:10])
````

## 2차점수(원리이해)
* 아웃라이어 제거 필수
* 정규화(스케일링)하고 점수보기

````python
# ================================ train,test 8:2 분리 =========================
y = train['count']
X = train.drop(['count','casual','registered'] , axis=1)
X80,X20,   y80,y20 =  train_test_split(X, y, test_size=0.2, random_state=4548)
print(X80.shape, X20.shape,   y80.shape,y20.shape)
````

````python
# ============================= 정규화(X) ====================
col_list = ['temp', 'atemp', 'humidity', 'windspeed']
scalerX = StandardScaler()       # (8708, 14)
scalerX.fit(X80[col_list])   
res80 = scalerX.transform(X80[col_list])
res20 = scalerX.transform(X20[col_list])
res80.shape , res20.shape
````

````python
y80.shape, np.array(y80).reshape(-1,1) 
````

````python
# =============== 정규화(y80: 0.32,0.23)   (pred: 0.32,0.23) ---원복--- (y20:100..200) ==============
scalerY = StandardScaler()        # (8708,   )
scalerY.fit(np.array(y80).reshape(-1,1) )   # 1차 --> 2차 
scalerY80 = scalerY.transform(np.array(y80).reshape(-1,1))

print( scalerY80.reshape(-1).shape  )
````

````python
scalerY80[:10]
````

````python
scalerX80 = X80[['season', 'holiday', 'workingday', 'weather','year','month', 'day', 'hour', 'weekday', 'day_type']].copy()
scalerX20 = X20[['season', 'holiday', 'workingday', 'weather','year','month', 'day', 'hour', 'weekday', 'day_type']].copy()

scalerX80[col_list] = res80
scalerX20[col_list] = res20
scalerX80.head(2)
````

````python
# ================================학습 & 평가 ==================================
lr = LinearRegression()
lr.fit(scalerX80, scalerY80)
scaler_pred = lr.predict(scalerX20)
pred = scalerY.inverse_transform(scaler_pred) #예측값원복

print(y20.shape, y20[:3])
print(scaler_pred[:3])
print(np.array(pred).reshape(-1).shape, pred[:3])
````

````python
mse  = mean_squared_error(y20, pred)
print(np.sqrt(mse))
````
