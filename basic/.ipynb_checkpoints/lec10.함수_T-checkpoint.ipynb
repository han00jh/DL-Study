# Lec10. 함수

## *args : tuple 파라미터

````python
def myprint(*bb):
    print(bb)
````

````python
myprint(10)
````

````python
def myprint(*args):
    print(args)
````

````python
myprint(10)
````

````python
myprint(1,2,3,4)
````

````python
myprint([1,2,3])
````

````python
def myprint(*args, num):
    print(args)
````

````python
#myprint(10)
````

````python
myprint(1,2,3, num=4)
````

````python
#myprint([1,2,3])
````

````python
def myprint(num, *args):
    print(args)
````

````python
#myprint(num=4 , 1,2,3)   에러에러
````

## **kargs : dic 파라미터

````python
def myprint(**bb):
    print(bb)
````

````python
#myprint({"aa" : 10})
#myprint("aa" = 10)
myprint(aa = 10)
````

````python
def myprint(**kargs):
    print(kargs)
````

````python
myprint(aa = 10)
````

````python
def myprint(htmlname, **kargs):
    print(htmlname)
    print(kargs)
````

````python
myprint("index.html", aa = 10)
````

````python
myprint("index.html", aa = 10, bb=20)
````

````python
def my_render_template(htmlname,  **kargs):
    print(htmlname)
    print(kargs)
````

````python
my_render_template('index.html'   
                           ,MY_KEY_MYLIST = 10
                           ,MY_KEY_CHART_DATA = [1,2,3]
                           )
````

# *args **kargs 혼합

````python
def myprint(htmlname,  *args , **kargs):
    print(args)
    print(htmlname)
    print(kargs)
````

````python
myprint("index.html",   10,20 ,     a=10,b=20)
````

````python
def myprint(htmlname,  *args , num, **kargs):
    print(args)
    print(htmlname)
    print(kargs)
````

````python
myprint("index.html",   10,20 ,    num=30,    a=10,b=20)
````
