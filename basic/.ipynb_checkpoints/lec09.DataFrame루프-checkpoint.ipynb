# lec09. DataFrame 루프

df.iterrows() vs. df.itertuples() vs. enumerate()
* <font size=4 color=red><b> iterrows() : 행을 시리즈로 반환 #인덱스로 뽑음
* <font size=4 color=red><b> itertuples() : 행반환  #자체 인덱스
* <font size=4 color=red><b> enumerate() : i번째, 행반환  #인덱스로 뽑음

```python
import numpy as np
import pandas as pd
```

```python
df =pd.read_csv("./datas/dept.csv")
```

## 컬럼 출력

```python
for r in df:
    print(r)
```

## iterrows()

```python
for idx, row in df.iterrows():#반복하면서 시리즈 형식으로 출력, 인덱스번호도 출력
    print(idx, row)
    print("--"*20)
```

## DF 값 출력

```python
print(df.columns.to_list())
```

```python
for r in df.values:
    print(r)
```

## itertuples()

```python
for  row in df.itertuples(): #인덱스 같이 뽑힘
    print(row, row.DNAME)
    print("--"*20)
```

## enumerate()

```python
for idx, row in enumerate(df): #인덱스에 컬럼도 붙어서 나옴
    print(idx, row)
    print("--"*20)
```

```python
for idx, row in enumerate(df.values): #반복되는 대상 넣고 인덱스를 같이 뺄 수 있다
    print(idx, row)
    print("--"*20)
```
