## for

## for + if

````python
for i in range(10) : 
    if(i%2 == 0):
        print(i)
````

````python
[  i for i in range(10) if(i%2 == 0)  ]
````

## if + else

````python
num = 10
if (num%2 == 0) :
    print("짝수")
else :
    print("홀수")
````

## for + for

````python
dan = [8,9]
gob = [1,2,3]
for d in dan:
    for g in gob:
        print(d*g)
````

````python
[ d*g for d in dan for g in gob ]
        
````

## 일반함수사용

* 성적 입력시 등급 출력

````python
def myscore(score):
    grade = ""
    if score>90:
        grade ="A"
    else :
        grade ="F"
    return grade
````

````python
myscore(99)
# 변수에 담에서 출력하기: grade =myscore(99)
#                       grade
````

## 람다 사용

````python
grade = lambda score : "A" if score > 90 else  "F"
grade (99)
````

````python
( lambda score : "A" if score > 90 else  "F") (99)
````

````python
def myscore(score):
    grade = ""
    if score>90:
        grade ="A"
    else :
        grade ="F"
    return grade
````

````python
(lambda score : myscore(score)) (99)
````

## DataFrame의 활용

````python
import pandas as pd
import numpy as np

df = pd.DataFrame(  { "name2":['kim','lee','lee','lee','lee']
                     , "age" : [10,20,np.nan,30,20]} )
df
````

````python
df.groupby('name2') ['age'].mean()
````

### <b>방법1) loc[ ]

````python
dfcp = df.copy()
````

````python
df.groupby('name2') ['age'].mean()
````

````python
dfcp.loc[ (df['name2']=='lee')    & (dfcp['age'].isna()==True) , "age" ] =23
dfcp
````

### <B>방법2) fillna(map())

````python
dfcp = df.copy()
````

````python
s = df.groupby('name2') ['age'].mean()
s
````

````python
dfcp['age'] = dfcp['age'].fillna(  dfcp['name2'].map( s.to_dict() ) )
dfcp
````

### <b>방법3) transform(lambda:_)

<font color= red> <b> dfcp['age'] = dfcp.groupby('name2')['age'].transform(lambda ag: ag.fillna(ag.mean()))

````python
dfcp = df.copy()
````

````python
lambda ag : ag.fillna(ag.mean())
````

````python
#dfcp.groupby('name2') ['age']  .mean()
# dfcp['age'].fillna(  dfcp['name2'].map( s ) )
dfcp['age'] = dfcp.groupby('name2') ['age'] .transform(lambda ag: ag.fillna(ag   .mean()))
dfcp
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame(  { 
                        "grade": ['A','A','A','B','B']
                      , "age" : [10,20,np.nan,30,20]      
                      , "score" : [100,90,80,70,80]        
                      , "name2": ['kim','lee','hong','park','jung']
                  } )
df
````

### [공통]  컬럼명을 메서드 밖에서 명시하면 결과는 같게 나온다
* score apply(), transform()

````python
#df.groupby("grade") ['score']  .mean()
#df.groupby("grade") ['score']  .apply( lambda sc : sc.mean())
df.groupby("grade") ['score']  .transform( lambda sc : sc.mean())
````

### [차이1] 컬럼명을 메서드 안에서 명시하면

* apply() : 그룹별 시리즈 리턴

````python
#df.groupby("grade") ['score']  .apply( lambda sc : sc.mean())
#df. groupby("grade")            .apply( lambda gdf: gdf['score'].mean())
df. groupby("grade")            .apply( lambda gdf: gdf['score'].mean(),include_groups=False)
````

* transform() : 에러

````python
#df. groupby("grade")            .transform( lambda gdf: gdf['score'].mean()) # transform은 ['score']안에두면 에러
df.groupby("grade") ['score']  .transform( lambda sc : sc.mean())
````

### [차이3] 연산 가능한 컬럼만 함수 적용한다면

````python
dfcp = df[['age','score']]
````

* apply() :그룹별 데이터프레임 출력

* transform() : (멀티인덱스)	기존 인덱스 유지

### [차이2]컬럼명을 메서드 안과 밖에 모두 명시하지 않는다면 -> 에러남 의미 없음

* apply() : 기존 데이터의 자료형과 함수가 연산 가능해야만 실행된다.

* transform() : 연산 가능한 컬럼만 알아서 필터링되어 결과가 나온다.  -> 지금은 에러남
