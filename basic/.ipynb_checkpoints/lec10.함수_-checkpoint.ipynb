## *args : tuple파라미터

````python
def myprint(*args):
    print(args)
````

````python
myprint(10)
````

````python
myprint(1,2,3,4)
````

````python
myprint([1,2,3])
````

````python
def myprint(*args, num):#순서
    print(args)
````

````python
#myprint(10)
````

````python
myprint(1,2,3, num=4)
````

````python
#myprint([1,2,3])
````

````python
def myprint(num, *args): 
    print(args)
````

## **kargs : dic파라미터

````python
def myprint(**bb):
    print(bb)
````

````python
myprint(aa =10)
````

````python
def myprint(**kargs):
    print(kargs)
````

````python
def myprint(htmlname, **kargs):#순서
    print(htmlname)
    print(kargs)
````

````python
myprint("index.html", aa =10)
````

## *args, **kargs 혼합

````python
def myprint(htmlname, *args, **kargs):#순서 일반파라미터 먼저, args, kargst순서. kargs는 제일 마지막. 뒤에 더 올수없음
    print(args)
    print(htmlname)
    print(kargs)
````

````python
myprint("index.html", 10,20,  a =10,b=20)
````
