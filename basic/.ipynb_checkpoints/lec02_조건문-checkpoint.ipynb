```python
#20대 남 군필 면제
age = 47
gen ="남"

if (20<=age<=29 and gen =="남"): #파이썬한정 가능문법ㅋㅋ (20<=age and age <=29)가 일반적
    print("군필")
else:
    print("면제")
```
