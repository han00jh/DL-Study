````python
model_list = [RandomForestClassifier(random_state=4567) ]

X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2,  random_state=4567, stratify=y)

for model in model_list:
    print(f" {model.__class__.__name__} --------------------" )  
    model.fit(X80, y80)
    pred = model.predict(X20)
    
    acc       = accuracy_score(y20, pred)
    f1        = f1_score(y20, pred, average='macro')
    
    report    = classification_report(y20, pred)
    cm        = confusion_matrix(y20, pred)

    print(f"acc:{acc:.4f}\t f1:{f1:.4f}")
    print(report)
    print(cm)
````

## KMeans
* https://www.linkedin.com/pulse/understanding-k-means-clustering-its-real-use-cases-security-arora

<pre>
Step-1: Select the value of K, to decide the number of clusters to be formed.
Step-2: Select random K points which will act as centroids.
Step-3: Assign each data point, based on their distance from the randomly selected points (Centroid), to the nearest/closest centroid which will form the predefined clusters.
Step-4: place a new centroid of each cluster.
Step-5: Repeat no.3, which reassign each datapoint to the new closest centroid of each cluster.
Step-6: If any reassignment occurs, then go to step-4 else go to Step 7.
Step-7: FINISH

### <font color=red size=4><b>실루엣 계수(Silhouette Coefficient)

* <b>군집이 얼마나 잘 되었는지를 평가하는 지표
* <b>군집내에서는 응집도가 좋아야 하고, 다른 군집과는 분리가 잘 되어 이었야 한다. <br><br> 
* <b>$ { coef = \frac{(b - a)} {max(a, b)} }$
    * <b> a : 군집내에서 헤더와의 평균 거리 (작으면 응집도가 좋다)
    * <b> b : 주변의 가장 가까운 군집과 헤더와의 평균 거리(다른 군집과 분리가 잘되있다)
-----------------------------------------------------------------------------------------
* <b>silhouette_samples(X, labels) : 모든 데이터들의 실루엣 계수
* <b>silhouette_score(X, labels) : silhouette_samples의 평균
    * -1 ~ 1 : 1에 가까울수록 군집이 잘 형성됨

``` python
sklearn.metrics.silhouette_score(X, labels, *, metric='euclidean', sample_size=None, random_state=None, **kwds)

sklearn.metrics.silhouette_samples(X, labels, *, metric='euclidean', **kwds)

````python
s_score = silhouette_score(X, model.labels_)
s_samples = silhouette_samples(X, model.labels_)
print(s_score)
print(np.mean(s_samples))
print(len(s_samples), len(X), s_samples)
````

### 2차 군집 : K=? k =n개

````python
inertia_list = []
for k in [2,3,4,5] :
    model = KMeans(n_clusters=k, random_state=5484)
    model.fit(X)
    df[f"km_cluster{k}"] = model.labels_

    #print(model.cluster_centers_)
    print(model.inertia_)
    inertia_list.append(model.inertia_)

    s_score = silhouette_score(X, model.labels_)
    print(f"실루엣 스코어: {s_score:.4f}")
    plt.figure(figsize=(10,2))
    
    plt.subplot(1,2,1)
    sns.scatterplot(x="pl", y="pw", data=df, hue=f"km_cluster{k}")
    sns.scatterplot(x=[4.95], y=[1.69],color = "red")
    sns.scatterplot(x=[1.56], y=[0.29],color = "red")
    
    plt.subplot(1,2,2)
    sns.scatterplot(x="sl", y="sw", data=df, hue=f"km_cluster{k}")

    plt.show()
    
````

### 3차 군집 + 시각화 + 결과분석

````python
X.head(2)
````

````python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.datasets import load_iris

for k in [2, 3, 4, 5]:
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 6)

    model = KMeans(n_clusters=k, random_state=5484)
    cluster_labels = model.fit_predict(X)

    s_score = silhouette_score(X, cluster_labels)
    print(f"n_clusters : {k} silhouette_score: {s_score}")
    s_samples = silhouette_samples(X, cluster_labels)

    # -------------------- Silhouette Plot --------------------
    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = s_samples[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / k)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i), fontsize=14)
        y_lower = y_upper + 10

    ax1.set_title("Silhouette plot")
    ax1.set_xlabel("coefficient")
    ax1.set_ylabel("cluster_label")
    ax1.axvline(x=s_score, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # -------------------- Scatter Plot --------------------
    colors = cm.nipy_spectral(cluster_labels.astype(float) / k)
    ax2.scatter(
        df["pl"], df["pw"],
        marker="o",
        s=80,                 # 점 크기 증가
        lw=0.5,
        alpha=0.8,
        c=colors,
        edgecolor="k"
    )
    
    centers = model.cluster_centers_

    ax2.scatter(
        centers[:, 2], #pl
        centers[:, 3], #pw
        marker="o",
        c="white",
        s=400,                # 중심점 크게
        alpha=1,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(
            c[2], c[3],
            marker=f"${i}$",
            s=300,            # 군집번호 크게
            alpha=1,
            edgecolor="k"
        )

    ax2.set_title("Cluster visualization")
    ax2.set_xlabel("pl")
    ax2.set_ylabel("pw")

    plt.suptitle(
        f"Silhouette analysis for KMeans (k={k} silhouette_score : {s_score :.4f})",
        fontsize=16,
        fontweight="bold",
    )

plt.show()
````

#### 3차 군집 + 시각화 + 결과분석

* 노이즈(-1) 검정색 처리

````python
from sklearn.cluster import DBSCAN
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np

for e in [0.3, 0.5, 0.8]:
    for s in [3, 5, 7]:

        model = DBSCAN(eps=e, min_samples=s)
        cluster_labels = model.fit_predict(X)

        unique_labels = set(cluster_labels)
        n_clusters = len(unique_labels - {-1})

        print(f"eps={e}, min_samples={s}, n_clusters={n_clusters}")


        # ------------------------ 차트 --------------------
        plt.figure(figsize=(8, 3))

        # 색상 지정 (-1은 검정)
        colors = [
            "pink" if label == -1
            else cm.nipy_spectral(label / max(n_clusters,1))
            for label in cluster_labels
        ]

        plt.scatter(
            df["pl"], df["pw"],
            c=colors,
            s=80,
            edgecolor="k",
            alpha=0.8
        )

        # 클러스터 평균 중심 표시
        for label in unique_labels:
            if label == -1:
                continue

            cluster_points = X[cluster_labels == label]
            center = cluster_points.mean(axis=0)

            plt.scatter(
                center[2], center[3],
                marker=f"${label}$",
                s=400,
                c="white",
                edgecolor="k"
            )

        plt.title(f"DBSCAN eps={e}, min_samples={s}")
        plt.xlabel("pl")
        plt.ylabel("pw")
        plt.show()
````

### 3차 군집 +시각화+결과분석

````python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.datasets import load_iris

for k in [2, 3, 4, 5]:
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 6)
    X=dfcp[['pc1', 'pc2']]

    model = KMeans(n_clusters=k, random_state=5484)
    cluster_labels = model.fit_predict(X)

    s_score = silhouette_score(X, cluster_labels)
    print(f"n_clusters : {k} silhouette_score: {s_score}")
    s_samples = silhouette_samples(X, cluster_labels)

    # -------------------- Silhouette Plot --------------------
    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = s_samples[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / k)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i), fontsize=14)
        y_lower = y_upper + 10

    ax1.set_title("Silhouette plot")
    ax1.set_xlabel("coefficient")
    ax1.set_ylabel("cluster_label")
    ax1.axvline(x=s_score, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # -------------------- Scatter Plot --------------------
    colors = cm.nipy_spectral(cluster_labels.astype(float) / k)
    ax2.scatter(
        dfcp["pc1"], dfcp["pc2"],
        marker="o",
        s=80,                 # 점 크기 증가
        lw=0.5,
        alpha=0.8,
        c=colors,
        edgecolor="k"
    )
    
    centers = model.cluster_centers_

    ax2.scatter(
        centers[:, 0], #pl
        centers[:, 1], #pw
        marker="o",
        c="white",
        s=400,                # 중심점 크게
        alpha=1,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(
            c[0], c[1],
            marker=f"${i}$",
            s=300,            # 군집번호 크게
            alpha=1,
            edgecolor="k"
        )

    ax2.set_title("Cluster visualization")
    ax2.set_xlabel("pc1")
    ax2.set_ylabel("pc2")

    plt.suptitle(
        f"Silhouette analysis for KMeans (k={k} silhouette_score : {s_score :.4f})",
        fontsize=16,
        fontweight="bold",
    )

plt.show()
````
