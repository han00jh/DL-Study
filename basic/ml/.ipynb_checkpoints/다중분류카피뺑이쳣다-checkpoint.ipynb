# 1차
* acc:0.6977 f1:0.4160

```python
model = RandomForestClassifier(random_state=5445)

X = wine.drop(['quality'], axis =1)
y = wine['quality']

X80, X20, y80, y20 = train_test_split(X, y, test_size = 0.2, random_state=5445, stratify=y)

model.fit(X80,y80)
pred = model.predict(X20)

acc = accuracy_score(y20, pred)
f1 = f1_score(y20, pred, average='macro') # average='binary' -> [None, 'micro'- 데이터 불균형시, 'macro-전체', 'weighted'].

print(f"acc:{acc:.4f} f1:{f1:.4f}")
```

# 2차
* Acc: 0.6962  F1 Score: 0.4173

```python
#type추가
red['type'] = 0
white['type'] = 1
wine = pd.concat([red, white], ignore_index=True)

#분리
X = wine.drop(['quality'], axis=1)
y = wine['quality']

# residual sugar, free sulfur dioxide, total sulfur dioxide, chlorides 로그변환
for col in ['residual sugar','free sulfur dioxide','total sulfur dioxide','chlorides']:
    X[col+'_log'] = np.log1p(X[col])

#피쳐
X['sugar_alcohol'] = X['residual sugar'] / X['alcohol'] # 단맛대비 알콜
X['sulfur_acidity'] = X['total sulfur dioxide'] / X['fixed acidity'] #방부제 대비 산도

X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2, random_state=5445, stratify=y)

model = RandomForestClassifier(
    n_estimators=200,         # 트리 개수
    random_state=5445,
    class_weight='balanced'   # 불균형 클래스 보정
)
model.fit(X80, y80)

pred = model.predict(X20)

acc = accuracy_score(y20, pred)
f1 = f1_score(y20, pred, average='macro')

print(f"Accuracy: {acc:.4f}")
print(f"Macro F1 Score: {f1:.4f}")
```

```python
#중요도 높은 순서부터 시각화
importances = model.feature_importances_
feat_names = X.columns

feat_imp = pd.DataFrame({'feature': feat_names, 'importance': importances})
feat_imp = feat_imp.sort_values(by='importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='importance', y='feature', data=feat_imp)
plt.title("RandomForest Feature Importance")
plt.show()
```

# 3차
* Acc: 0.6377 F1 Score: 0.4003

```python
#type추가
red['type'] = 0
white['type'] = 1
wine = pd.concat([red, white], ignore_index=True)

#분리
X = wine.drop(['quality'], axis=1)
y = wine['quality']

# residual sugar, free sulfur dioxide, total sulfur dioxide, chlorides 로그변환
for col in ['residual sugar','free sulfur dioxide','total sulfur dioxide','chlorides']:
    X[col+'_log'] = np.log1p(X[col])

#피쳐
X['sugar_alcohol'] = X['residual sugar'] / X['alcohol'] # 단맛대비 알콜
X['sulfur_acidity'] = X['total sulfur dioxide'] / X['fixed acidity'] #방부제 대비 산도
X['acid_alc_ratio'] = X['volatile acidity'] / (X['alcohol']+1e-5) # 산화성 대비 알콜
X['den_alc_ratio'] = X['alcohol'] / X['density'] # 알콜 대비 밀도

X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2, random_state=5445, stratify=y)

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=15,
    min_samples_leaf=3,
    random_state=5445,
    class_weight='balanced')
model.fit(X80, y80)

pred = model.predict(X20)

acc = accuracy_score(y20, pred)
f1 = f1_score(y20, pred, average='macro')

print(f"Accuracy: {acc:.4f}")
print(f"Macro F1 Score: {f1:.4f}")
```

# 4차
* Accuracy: 0.6231 F1 Score: 0.3259

```python
#type추가
red['type'] = 0
white['type'] = 1
wine = pd.concat([red, white], ignore_index=True)

#분리
X = wine.drop(['quality'], axis=1)
y = wine['quality']

# residual sugar, free sulfur dioxide, total sulfur dioxide, chlorides 로그변환
for col in ['residual sugar','free sulfur dioxide','total sulfur dioxide','chlorides']:
    X[col+'_log'] = np.log1p(X[col])

#피쳐
X['sugar_alcohol'] = X['residual sugar'] / X['alcohol'] # 단맛대비 알콜
X['sulfur_acidity'] = X['total sulfur dioxide'] / X['fixed acidity'] #방부제 대비 산도
X['acid_alc_ratio'] = X['volatile acidity'] / (X['alcohol']+1e-5) # 산화성 대비 알콜
X['den_alc_ratio'] = X['alcohol'] / X['density'] # 알콜 대비 밀도


X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2, random_state=5445, stratify=y)

#회귀 가중치 직접입력
class_weights = {3:3, 4:2, 5:1, 6:1, 7:2, 8:3}

model = RandomForestClassifier(
    n_estimators=300,  # 트리 개수
    max_depth=12, #과적합방지
    min_samples_leaf=5, #회귀 안정화
    random_state=5445,
    class_weight= class_weights)
model.fit(X80, y80)

pred = model.predict(X20)

acc = accuracy_score(y20, pred)
f1 = f1_score(y20, pred, average='macro')

print(f"Accuracy: {acc:.4f}")
print(f"Macro F1 Score: {f1:.4f}")
```
