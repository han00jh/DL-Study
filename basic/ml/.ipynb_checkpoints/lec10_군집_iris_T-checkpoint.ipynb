# <b> 분류 모델 점수

````python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report,          confusion_matrix
from sklearn.metrics import accuracy_score,      f1_score
````

````python
X80,X20, y80, y20 = train_test_split(X, y, test_size=0.2,random_state=4878, shuffle=True, stratify=y)
model = RandomForestClassifier(random_state=5787)
model.fit(X80, y80)
pred = model.predict(X20)

acc = accuracy_score(y20, pred)
f1  = f1_score(y20, pred, average="macro")
print(f"acc: {acc:.4f}  f1 : {f1:.4f}") 

report = classification_report(y20, pred)
cm = confusion_matrix(y20, pred)
print(report)
print(cm)
````

## KMeans
* https://www.linkedin.com/pulse/understanding-k-means-clustering-its-real-use-cases-security-arora

<pre>
Step-1: Select the value of K, to decide the number of clusters to be formed.
Step-2: Select random K points which will act as centroids.
Step-3: Assign each data point, based on their distance from the randomly selected points (Centroid), to the nearest/closest centroid which will form the predefined clusters.
Step-4: place a new centroid of each cluster.
Step-5: Repeat no.3, which reassign each datapoint to the new closest centroid of each cluster.
Step-6: If any reassignment occurs, then go to step-4 else go to Step 7.
Step-7: FINISH

### 1차 군집 (K=3)

````python
model = KMeans(n_clusters=3, random_state=5484)
model.fit(X)
model.labels_
````

````python
df['km_cluster'] = model.labels_
````

````python
plt.figure(figsize=(10,4))

plt.subplot(2,2,1)
sns.scatterplot(x="pl", y="pw", data=df, hue="km_cluster")

plt.subplot(2,2,2)
sns.scatterplot(x="sl", y="sw", data=df, hue="km_cluster")

plt.subplot(2,2,3)
sns.scatterplot(x="pl", y="pw", data=df, hue="target")

plt.subplot(2,2,4)
sns.scatterplot(x="sl", y="sw", data=df, hue="target")


plt.show()
````

### <font color=red size=4><b>실루엣 계수(Silhouette Coefficient)

* <b>군집이 얼마나 잘 되었는지를 평가하는 지표
* <b>군집내에서는 응집도가 좋아야 하고, 다른 군집과는 분리가 잘 되어 이었야 한다. <br><br> 
* <b>$ { coef = \frac{(b - a)} {max(a, b)} }$
    * a : 군집내에서 헤더와의 평균 거리 (작으면 응집도가 좋다)
    * b : 주변의 가장 가까운 군집과 헤더와의 평균 거리(다른 군집과 분리가 잘되있다)
----
* <b>silhouette_samples(X, labels) : 모든 데이터들의 실루엣 계수 150개
* <b>silhouette_score(X, labels) : silhouette_samples(실루엣 계수 150개 평균)
    * -1 ~ 1 : 1에 가까울수록 군집이 잘되있다 

```python
sklearn.metrics.silhouette_score(X, labels, *, metric='euclidean', sample_size=None, random_state=None, **kwds)

sklearn.metrics.silhouette_samples(X, labels, *, metric='euclidean', **kwds)
                                 

````python
s_score   = silhouette_score(X, model.labels_)
s_samples = silhouette_samples(X, model.labels_)  
print(s_score)
print(np.mean(s_samples))
print(len(s_samples),   len(X), s_samples  )
````

### 2차 군집 (K=2,3,4,5)

````python
inertia_list = []
for  k in [2,3,4,5] : 
    print(X.head(2))
    model = KMeans(n_clusters=k, random_state=5484)
    model.fit(X)
    df[f"km_cluster{k}"] = model.labels_

    #print(model.cluster_centers_)
    print(model.inertia_)  #Sum of squared distances of samples to their closest cluster center
    inertia_list.append(model.inertia_)

    s_score   = silhouette_score(X, model.labels_)
    print(f"실루엣스코어 : {s_score:.4f}")

    
    plt.figure(figsize=(10,2))
    plt.subplot(1,2,1)
    sns.scatterplot(x="pl", y="pw", data=df, hue=f"km_cluster{k}")
    sns.scatterplot(x=[4.95], y=[1.69], color="red")
    sns.scatterplot(x=[1.56], y=[0.29], color="red")
    
    plt.subplot(1,2,2)
    sns.scatterplot(x="sl", y="sw", data=df, hue=f"km_cluster{k}")
    plt.show()
    
````

````python
plt.figure(figsize=(4,2))
plt.plot( [2,3,4,5] , inertia_list)
plt.title("엘보차트(적정K)")
plt.show()
````
