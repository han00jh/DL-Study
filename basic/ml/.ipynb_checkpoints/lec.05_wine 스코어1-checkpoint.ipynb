# <b> 2. EDA

## 1. 타겟 변수 생성 및 클래스 분포 확인

```python
# 1. 타겟 생성 (기존 코드 유지)
def categorize_quality(q):
    if q <= 4: return 'Low'
    elif q <= 7: return 'Medium'
    else: return 'High'

wine['target'] = wine['quality'].apply(categorize_quality)
wine['target'] = pd.Categorical(wine['target'], categories=['Low', 'Medium', 'High'], ordered=True)

# 2. 오류 해결 핵심: 숫자를 문자로 바꾼 임시 컬럼 생성
wine['type_name'] = wine['type'].map({0: 'red', 1: 'white'})

# 3. 시각화 (hue를 'type' 대신 'type_name'으로 설정)
plt.figure(figsize=(10, 5))
sns.countplot(data=wine, x='target', hue='type_name', palette={'red': '#8E0808', 'white': '#F2E8C4'})
plt.title('Quality Class Distribution by Wine Type')
plt.show() ;

# 실제 수치 확인
print(wine.groupby(['type_name', 'target'], observed=True).size().unstack(fill_value=0))
```

## 2. 피처별 변별력 확인 (Boxplot)

```python
# 분석하고 싶은 주요 피처들
features = ['alcohol', 'volatile acidity', 'sulphates', 'citric acid']

plt.figure(figsize=(15, 10))
for i, col in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=wine, x='target', y=col, hue='type', palette='Set2')
    plt.title(f'{col.capitalize()} vs Quality Class')

plt.tight_layout()
plt.show() ;
```

## 3. 상관관계 분석 (Correlation Heatmap)

```python
plt.figure(figsize=(12, 5))

corr_matrix = wine.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", mask=mask, cmap='coolwarm')    
plt.show() ;
```

## 4. 타입별 성분 분포 분석 (KDE Plot)

```python
# 분석할 주요 피처 선택
analysis_features = ['fixed acidity', 'volatile acidity', 'residual sugar', 'total sulfur dioxide', 'alcohol']

plt.figure(figsize=(15, 10))

for i, col in enumerate(analysis_features, 1):
    plt.subplot(3, 2, i)
    # 레드(0)와 화이트(1)의 분포 비교
    sns.kdeplot(data=wine[wine['type'] == 0], x=col, fill=True, color='#8E0808', label='Red')
    sns.kdeplot(data=wine[wine['type'] == 1], x=col, fill=True, color='#F2E8C4', label='White')
    plt.title(f'Distribution of {col} by Wine Type')
    plt.legend()

plt.tight_layout()
plt.show() ;
```

# <b> 3. 전처리 및 가공

## 1. 데이터 분리 (Red & White)

```python
# 1. 타입별 데이터 분리 (전체 wine -> red, white)
# EDA 결과에 따라 0이 레드, 1이 화이트라고 가정합니다.
red = wine[wine['type'] == 0].copy()
white = wine[wine['type'] == 1].copy()

# 2. 학습에 불필요한 컬럼 정의
# target의 원본인 quality와 시각화용 label/name 컬럼들을 제거 대상에 넣습니다.
drop_cols = ['type', 'quality', 'target'] 
if 'type_label' in wine.columns: drop_cols.append('type_label')
if 'type_name' in wine.columns: drop_cols.append('type_name')

# 3. Feature(X)와 Label(y) 분리
red_X = red.drop(columns=drop_cols)
red_y = red['target']

white_X = white.drop(columns=drop_cols)
white_y = white['target']

print(f"Red 데이터 갯수: {len(red_X)}")
print(f"White 데이터 갯수: {len(white_X)}")
```

## 2. 피처 선택 및 가중치 설정

```python
# 1. 제외할 피처 목록 (EDA 상관계수가 낮았던 항목들)
# 예시입니다. 직접 보신 히트맵에서 수치가 0.05 미만인 것들을 넣어주세요.
features_to_drop = ['pH', 'free sulfur dioxide'] 

# 2. Red와 White에서 각각 해당 피처 제거
red_X = red_X.drop(columns=features_to_drop)
white_X = white_X.drop(columns=features_to_drop)

# 3. 중요 피처 가중치 강조 (선택 사항)
# 모델이 Alcohol이나 Volatile Acidity를 더 중요하게 보게 하려면 
# 지금은 컬럼을 유지하는 것만으로도 충분합니다. 
# (나중에 학습 시 파라미터로 조절하는 것이 더 정확합니다.)

print(f"남은 피처 목록: {list(red_X.columns)}")
```

## 3. 결측치 확인 및 처리

```python
# 결측치가 있는지 확인 (0이 나와야 정상입니다)
print("Red 결측치: ", red_X.isnull().sum().sum())
print("White 결측치: ", white_X.isnull().sum().sum())
```

## 4. 이상치(Outlier) 처리

```python
def handle_outliers(df):
    df_clean = df.copy()
    for col in df_clean.columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # 이상치를 상/하한값으로 보정 (삭제 대신 보정을 택해 데이터 손실 방지)
        df_clean[col] = df_clean[col].clip(lower_bound, upper_bound)
    return df_clean

# Red와 White에 각각 적용
red_X = handle_outliers(red_X)
white_X = handle_outliers(white_X)

print("이상치 보정 완료")
```

## 5. 피처 스케일링 (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler_red = StandardScaler()
scaler_white = StandardScaler()

# 각 타입에 맞춰 스케일링 (타입별로 분포가 다르므로 따로 학습시킵니다)
red_X_scaled = scaler_red.fit_transform(red_X)
white_X_scaled = scaler_white.fit_transform(white_X)

# 다시 DataFrame 형태로 변환 (컬럼명 유지를 위해)
red_X = pd.DataFrame(red_X_scaled, columns=red_X.columns)
white_X = pd.DataFrame(white_X_scaled, columns=white_X.columns)
```

## 6. 클래스 불균형 해소 (SMOTE)

```python
from imblearn.over_sampling import SMOTE

# 1. SMOTE 객체 생성 (재현성을 위해 random_state 설정)
smote = SMOTE(random_state=42)

# 2. Red 와인 데이터 증식
# 주의: y 데이터가 문자열('Low' 등)이어도 SMOTE는 내부적으로 처리해줍니다.
red_X_res, red_y_res = smote.fit_resample(red_X, red_y)

# 3. White 와인 데이터 증식
white_X_res, white_y_res = smote.fit_resample(white_X, white_y)

# 4. 결과 확인
print("--- Red Wine Class Distribution ---")
print(f"Before: {red_y.value_counts().to_dict()}")
print(f"After SMOTE: {red_y_res.value_counts().to_dict()}")

print("\n--- White Wine Class Distribution ---")
print(f"Before: {white_y.value_counts().to_dict()}")
print(f"After SMOTE: {white_y_res.value_counts().to_dict()}")
```

# <b> 4. 모델 학습

## 1. 타입별 최적 모델 탐색 코드 (Red & White 독립 실행)

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import logging
logging.getLogger('lightgbm').setLevel(logging.ERROR)
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, classification_report

# 1. 모델과 파라미터 그리드 설정
model_params = {
    'rf': {
        'model': RandomForestClassifier(random_state=42, class_weight='balanced'),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 20, None],
            'min_samples_split': [2, 5]
        }
    },
    'xgb': {
        'model': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 6, 9]
        }
    },
    'lgbm': {
        'model': LGBMClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'num_leaves': [31, 50]
        }
    }
}

# target이 문자열('Low' 등)이므로 학습을 위해 숫자로 임시 변환 (XGB, LGBM 용)
target_map = {'Low': 0, 'Medium': 1, 'High': 2}
inv_target_map = {v: k for k, v in target_map.items()}

def find_best_model(X, y, wine_type):
    print(f"\n{'='*20} {wine_type.upper()} WINE MODEL SEARCH {'='*20}")
    y_encoded = y.map(target_map)
    scores = []

    for model_name, mp in model_params.items():
        grid = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='f1_macro', n_jobs=-1)
        grid.fit(X, y_encoded)
        scores.append({
            'model': model_name,
            'best_score': grid.best_score_,
            'best_params': grid.best_params_
        })
        print(f"{model_name} 완료 (Best F1: {grid.best_score_:.4f})")

    best_res = sorted(scores, key=lambda x: x['best_score'], reverse=True)[0]
    return best_res

# 2. 실행 (Red 와인과 White 와인 각각 최적 모델 찾기)
best_red_model = find_best_model(red_X_res, red_y_res, "red")
best_white_model = find_best_model(white_X_res, white_y_res, "white")

print("\n\n[최종 결과 요약]")
print(f"Red 와인 최적 모델: {best_red_model['model']} (F1: {best_red_model['best_score']:.4f})")
print(f"White 와인 최적 모델: {best_white_model['model']} (F1: {best_white_model['best_score']:.4f})")
```

## 2. 모델 확정 및 피처 중요도(Feature Importance) 분석

```python
import matplotlib.pyplot as plt
import seaborn as sns

def finalize_and_plot(X, y, best_params, wine_type):
    # 1. 최적의 파라미터로 최종 모델 학습
    final_model = LGBMClassifier(**best_params, random_state=42, verbosity=-1)
    # y값 인코딩 (Low:0, Medium:1, High:2)
    y_encoded = y.map({'Low': 0, 'Medium': 1, 'High': 2})
    final_model.fit(X, y_encoded)
    
    # 2. 피처 중요도 추출
    feature_imp = pd.DataFrame({
        'Value': final_model.feature_importances_,
        'Feature': X.columns
    }).sort_values(by='Value', ascending=False)
    
    # 3. 시각화
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Value", y="Feature", data=feature_imp, palette='viridis')
    plt.title(f'{wine_type.upper()} Wine - Feature Importance (LGBM)')
    plt.tight_layout()
    plt.show()
    
    return final_model

# Red와 White 각각 최종 모델 확정 및 중요도 확인
print("최종 모델 확정 및 피처 분석 중...")
red_model = finalize_and_plot(red_X_res, red_y_res, best_red_model['best_params'], "red")
white_model = finalize_and_plot(white_X_res, white_y_res, best_white_model['best_params'], "white")
```

## 3. 최종 마무리: 혼동 행렬 (Confusion Matrix)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(X, y, model, wine_type):
    y_encoded = y.map({'Low': 0, 'Medium': 1, 'High': 2})
    y_pred = model.predict(X)
    
    cm = confusion_matrix(y_encoded, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'Medium', 'High'])
    
    plt.figure(figsize=(6, 6))
    disp.plot(cmap='Blues', values_format='d')
    plt.title(f'{wine_type.upper()} Wine Confusion Matrix')
    plt.show() ;

# 최종 검증 실행
plot_confusion_matrix(red_X_res, red_y_res, red_model, "red")
plot_confusion_matrix(white_X_res, white_y_res, white_model, "white")
```

# <b> 5. 평가 및 검증

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def evaluate_model(X, y, model, wine_type):
    # 1. 예측값 생성
    # SMOTE된 데이터(res)가 아닌, 검증용 데이터를 넣어야 정확한 평가가 가능합니다.
    y_encoded = y.map({'Low': 0, 'Medium': 1, 'High': 2})
    y_pred = model.predict(X)
    
    # 2. 상세 리포트 출력 (F1-score 확인)
    print(f"\n{'='*20} {wine_type.upper()} WINE FINAL REPORT {'='*20}")
    print(classification_report(y_encoded, y_pred, target_names=['Low', 'Medium', 'High']))
    
    # 3. 혼동 행렬 시각화
    fig, ax = plt.subplots(figsize=(6, 6))
    cm = confusion_matrix(y_encoded, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'Medium', 'High'])
    disp.plot(cmap='Purples', ax=ax, values_format='d')
    plt.title(f'{wine_type.upper()} Wine Confusion Matrix')
    plt.show()

# 최종 검증 실행
evaluate_model(red_X, red_y, red_model, "red")
evaluate_model(white_X, white_y, white_model, "white")
```
