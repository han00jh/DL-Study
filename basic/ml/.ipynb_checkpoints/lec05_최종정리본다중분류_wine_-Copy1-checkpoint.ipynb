# 라벨링

* 하나하나 나눠서 했을 때 점수 넘 낮길래 걍 품질 나눠서함

```python
def localize_quality(quality): #품질 
    if quality <= 5: 
        return 0 #낮음 
    elif quality  == 6: 
        return 1 #중간 
    else: 
        return 2 #높음
wine['quality'] = wine['quality'].apply(localize_quality) 
print(wine['quality'].value_counts().sort_index())
```

##  히스토그램

| 항목 | 영문 명칭 | 특징 및 요약 |
| :--- | :--- | :--- |
| **고정 산도** | Fixed Acidity | 비휘발성 산, 와인의 기본 신맛 구조 형성 |
| **휘발성 산도** | Volatile Acidity | 아세트산 수치, 과다 시 식초 맛 등 품질 저하 유발 |
| **구연산** | Citric Acid | 소량 함유, 와인의 신선함과 풍미 부여 |
| **잔류 당분** | Residual Sugar | 발효 후 남은 당, 45g/L 초과 시 스위트 와인 분류 |
| **염화물** | Chlorides | 와인 내 염분(소금) 함량 |
| **유리 이산화황** | Free Sulfur Dioxide | 가스/이온 평형 상태, 미생물 억제 및 산화 방지 |
| **총 이산화황** | Total Sulfur Dioxide | 유리+결합 이산화황, 50ppm 초과 시 불쾌한 향 발생 |
| **밀도** | Density | 알코올 함량 시 감소, 당분 함량 시 증가 |
| **pH** | pH | 산성도(0 ~ 14), 와인은 통상 3 ~4 pH 유지 |
| **황산염** | Sulphates | $SO_2$ 수치 유도 첨가제, 항균 및 항산화 작용 |

* 레드랑 화이트는 화학적 조성이 다름
* 그래서 각각이랑 합쳤을때해봄

```python
def myscore_scaled(df):
    model = RandomForestClassifier(random_state=5445)
    scaler = StandardScaler()

    df_new = df.copy()

    # 1. 알코올과 밀도 조합 (품질과 가장 상관관계 높음)
    df_new['alc_density_ratio'] = df_new['alcohol'] / df_new['density']
    # 2. 산성도 합계
    df_new['total_acidity'] = df_new['fixed acidity'] + df_new['volatile acidity']
    # 3. 설탕과 산미의 비율
    df_new['sugar_acid_ratio'] = df_new['residual sugar'] / (df_new['citric acid'] + 0.001) # 0 나누기 방지
    

    X = df_new.drop(['quality','type'], axis = 1)
    y = df_new['quality']

    X80, X20, y80, y20 = train_test_split(X, y, test_size = 0.2, random_state=5445, stratify=y)

    X80_scaled = scaler.fit_transform(X80)
    X20_scaled = scaler.transform(X20)

    model = RandomForestClassifier(
    random_state=5445, 
    class_weight='balanced')

    model.fit(X80_scaled, y80)
    pred = model.predict(X20_scaled)

    acc = accuracy_score(y20, pred)
    f1 = f1_score(y20, pred, average='micro') 

    print(f"Scaled Acc:{acc:.4f} Scaled F1:{f1:.4f}")
    print(classification_report(y20, pred, target_names=['Low', 'Medium', 'High']))
    return y20, pred





# 1. 원본 데이터에서 레드(0)와 화이트(1)를 각각 추출
red_wine = wine[wine['type'] == 0].copy()
white_wine = wine[wine['type'] == 1].copy()

# 2. 레드와인 성적표 뽑기
print("="*30)
print("   [RED WINE ONLY SCORE]   ")
print("="*30)
myscore_scaled(red_wine)

# 3. 화이트와인 성적표 뽑기
print("\n" + "="*30)
print("  [WHITE WINE ONLY SCORE]  ")
print("="*30)
myscore_scaled(white_wine)

# 4. 와인 성적표 뽑기
print("\n" + "="*30)
print("  [WINE SCORE]  ")
print("="*30)
myscore_scaled(wine)
```

```python
y20, pred = myscore_scaled(wine)
```

```python
# 클래스별 F1-score 시각화
from sklearn.metrics import classification_report

report = classification_report(y20, pred, target_names=['Low', 'Medium', 'High'], output_dict=True)
report_df = pd.DataFrame(report).transpose()

plt.figure(figsize=(8, 5))
sns.barplot(x=report_df.index[:3], y=report_df['f1-score'][:3], palette='magma')
plt.title('F1-Score by Quality Group')
plt.ylabel('F1-Score')
plt.ylim(0, 1)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, f1_score, accuracy_score

def myscore_master_f1(df):
    # 1. 파생 변수 생성 (모델에게 힌트 주기)
    df_new = df.copy()
    df_new['alc_density_ratio'] = df_new['alcohol'] / (df_new['density'] + 0.001) #산도지수
    df_new['acid_ratio'] = df_new['volatile acidity'] / (df_new['fixed acidity'] + 0.001) #보존상태 
    df_new['sugar_acid_ratio'] = df_new['residual sugar'] / (df_new['citric acid'] + 0.001) #바디감(알콜,당분)

    X = df_new.drop(['quality', 'type'], axis=1)
    y = df_new['quality']

    X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2, random_state=5445, stratify=y)

    # 2. 파라미터 그리드 (나무의 깊이와 개수를 조절)
    param_grid = {
        'n_estimators': [300, 500],
        'max_depth': [20, 30, None],
        'min_samples_split': [2, 5],
        'class_weight': ['balanced'] # 희귀 클래스 가중치 부여
    }

    # 3. GridSearchCV 설정 (scoring을 'f1_macro'로 설정하여 모든 클래스 성능 중시)
    grid_search = GridSearchCV(
        RandomForestClassifier(random_state=5445), 
        param_grid, 
        cv=5, 
        n_jobs=-1, 
        scoring='f1_macro' 
    )

    grid_search.fit(X80, y80)

    # 4. 결과 출력
    best_model = grid_search.best_estimator_
    pred = best_model.predict(X20)

    acc = accuracy_score(y20, pred)
    f1 = f1_score(y20, pred, average='macro') # 전체 클래스의 평균 F1

    print(f"최적 파라미터: {grid_search.best_params_}")
    print(f"7-Class Acc: {acc:.4f}")
    print(f"7-Class F1 (Macro): {f1:.4f}")
    
    # 클래스별 상세 점수 확인 (어떤 점수를 못 맞추는지 한눈에 보임)
    print("\n[상세 성적표]")
    print(classification_report(y20, pred))
    
    return y20, pred

# 실행
y20, pred = myscore_master_f1(wine)
```

```python
def myscore_angry_ver(df):
    df_new = df.copy()
    
    # [가공 1] 산도 시너지 피처 
    df_new['total_acid'] = df_new['fixed acidity'] + df_new['volatile acidity'] + df_new['citric acid']
    df_new['acid_ph_ratio'] = df_new['total_acid'] / (df_new['pH'] + 0.001)
    
    # [가공 2] 비율 (산도)
    df_new['so2_ratio'] = df_new['free sulfur dioxide'] / (df_new['total sulfur dioxide'] + 0.001)
    
    # [가공 3] 알콜, 당분의 상관관계 (바디감 지수)
    df_new['alcohol_sugar_ratio'] = df_new['alcohol'] * df_new['residual sugar']
    
    X = df_new.drop(['quality', 'type'], axis=1)
    y = df_new['quality']

    X80, X20, y80, y20 = train_test_split(X, y, test_size=0.2, random_state=5445, stratify=y)

    # [LGBM 튜닝] 7단계 분류에서 '디테일'을 살리는 파라미터
    model = LGBMClassifier(
        n_estimators=1500,        
        learning_rate=0.01,     
        num_leaves=255,          
        min_child_samples=10, 
        class_weight='balanced',
        random_state=5445,
        verbosity=-1
    )

    model.fit(X80, y80)
    pred = model.predict(X20)

    print(f" FINAL Micro: {f1_score(y20, pred, average='micro'):.4f}")
    print(f" FINAL Macro: {f1_score(y20, pred, average='macro'):.4f}")
    
    return y20, pred
```

```python
myscore_angry_ver(wine)
```
