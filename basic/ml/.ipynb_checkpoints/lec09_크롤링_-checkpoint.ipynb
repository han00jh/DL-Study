# <b>[실습] ytn 뉴스
* https://www.ytn.co.kr/news/list.php?mcd=0107

## 크롤링

````python
import requests
res = requests.get("https://www.ytn.co.kr/news/list.php?mcd=0107")
html_status = res.status_code
html_doc = res.text
print(html_status , html_doc[:200])
````

````python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')
soup.title.string
````

````python
for row in soup.find_all('a')[:10]  :
    print(row.get('href'))
````

````python
news_list = soup.select("#container > div > div.content > div > div.news_list_wrap > div")
len(news_list)
````

``` html
#container > div > div.content > div > div.news_list_wrap > div
#container > div > div.content > div > div.news_list_wrap > div:nth-child(1) > div.text_area > div.title > a
#container > div > div.content > div > div.news_list_wrap > div:nth-child(1) > div.text_area > div.info > div.date
#container > div > div.content > div > div.news_list_wrap > div:nth-child(1) > div.photo > a > img

## DataFrame 으로 변환

````python
import numpy as np
import pandas as pd
````

````python
temp_list = []
for div_news in news_list:
    href  = div_news.select_one("div.text_area > div.title > a").get("href")
    title = div_news.select_one("div.text_area > div.title > a").text
    rdate = div_news.select_one("div.text_area > div.info > div.date").text
    img   = div_news.select_one("div.photo > a > img").get("data-src")
    temp_list.append(  [href, title, rdate, img] )
    #print(href, img, rdate, title)
df = pd.DataFrame(temp_list, columns=["href","title","rdate","img"])
df.head(2)
````

## DB에 넣기
* 테이블명 : ytn_news

````python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine("oracle+cx_oracle://it:0000@localhost:1521/xe")
conn = engine.connect()
````

````python
df.to_sql(
    name="ytn_news",
    con=engine,
    if_exists="append",   # append | replace | fail
    index=False
)
````

* 확인용

````python
#df = pd.read_csv("datas/ddd.csv")
redb = pd.read_sql("SELECT * FROM ytn_news", conn)
redb.head(2)
````
