### 이상치 처리

```python
wine
```

```python
import numpy as np
import pandas as pd

# 1. 대상 컬럼 리스트 (순수 수치형 피처)
numeric_features = [
    'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
]

# 2. 로그 변환 (데이터 분포 안정화 및 log_features 생성)
log_features = []
for col in numeric_features:
    col_name = f'log_{col}'
    wine[col_name] = np.log1p(wine[col])
    log_features.append(col_name)

# 3. 이상치 대치 작업 (타입별 + 퀄리티별 경계값 대치)
summary_data = []

# 타입(Red/White)과 퀄리티(4~7)를 순회하며 정밀하게 처리
for t_val in [0, 1]:
    type_name = 'Red' if t_val == 0 else 'White'
    
    for q_val in sorted(wine['quality'].unique()):
        # 3, 8, 9 등급은 데이터의 희소성을 보존하기 위해 처리 제외
        if q_val in [3, 8, 9]: continue
        
        mask = (wine['type'] == t_val) & (wine['quality'] == q_val)
        if wine[mask].empty: continue
        
        for col in log_features:
            # 해당 타입 & 퀄리티 그룹 내에서의 IQR 계산
            Q1 = wine.loc[mask, col].quantile(0.25)
            Q3 = wine.loc[mask, col].quantile(0.75)
            IQR = Q3 - Q1
            
            # 3.0 IQR 기준 (극단적 이상치만 타겟팅)
            lower_limit = Q1 - 3.0 * IQR
            upper_limit = Q3 + 3.0 * IQR
            
            # [핵심] 이상치를 중앙값이 아닌 정상 범위의 끝단(경계값)으로 밀어 넣음
            outliers_lower = wine[mask & (wine[col] < lower_limit)].index
            outliers_upper = wine[mask & (wine[col] > upper_limit)].index
            
            wine.loc[outliers_lower, col] = lower_limit
            wine.loc[outliers_upper, col] = upper_limit
            
            if len(outliers_lower) + len(outliers_upper) > 0:
                summary_data.append({
                    'Type': type_name,
                    'Quality': q_val,
                    'Column': col,
                    'Lower_Fixed': len(outliers_lower),
                    'Upper_Fixed': len(outliers_upper)
                })

# 결과 확인
print("=== 🍷 [EDA 직후] 퀄리티별 경계값(Limit) 기반 대치 요약 ===")
if summary_data:
    print(pd.DataFrame(summary_data).head(10))
else:
    print("3.0 IQR 기준을 벗어나는 극단적 이상치가 발견되지 않았습니다.")
```

### <b> 품질별 경계값 클리핑 (Quality-based Boundary Clipping)

```python
# 1. 대상 컬럼 및 로그 변환 수행 (리스트 정의 포함)
numeric_features = [
    'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
]

log_features = []
for col in numeric_features:
    col_name = f'log_{col}'
    wine[col_name] = np.log1p(wine[col])
    log_features.append(col_name)

# 2. 타입별 & 품질별 경계값 대치 (Clipping) [cite: 2026-02-03]
summary_data = []

for t_val in [0, 1]: # 0: Red, 1: White
    for q_val in sorted(wine['quality'].unique()):
        # 데이터가 부족한 희귀 등급(3, 8, 9)은 보존하여 특이 패턴 유지 [cite: 2026-02-03]
        if q_val in [3, 8, 9]: continue
        
        mask = (wine['type'] == t_val) & (wine['quality'] == q_val)
        if wine[mask].empty: continue
        
        for col in log_features:
            Q1 = wine.loc[mask, col].quantile(0.25)
            Q3 = wine.loc[mask, col].quantile(0.75)
            IQR = Q3 - Q1
            
            # 3.0 IQR 밖의 극단적 수치만 정상 범위 끝단으로 조정 [cite: 2026-02-04]
            lower_limit = Q1 - 3.0 * IQR
            upper_limit = Q3 + 3.0 * IQR
            
            wine.loc[mask & (wine[col] < lower_limit), col] = lower_limit
            wine.loc[mask & (wine[col] > upper_limit), col] = upper_limit
            
            # 처리 이력 기록
            outliers_count = len(wine[mask & ((wine[col] < lower_limit) | (wine[col] > upper_limit))])
            if outliers_count > 0:
                summary_data.append({'Type': t_val, 'Quality': q_val, 'Col': col, 'Fixed': outliers_count})

print("✅ 퀄리티별 경계값 대치 작업 완료")
```

## 2. 파생피쳐 생성

```python
# 2. [근거: acidity_stability]
# fixed acidity와 pH의 불일치 구간 확인 -> "개별 지표의 한계 증명"
# ---------------------------------------------------------
sns.scatterplot(data=wine, x='fixed acidity', y='pH', hue='quality', palette='coolwarm', alpha=0.5)
plt.title('2. Acidity vs pH (Chemical Stability)')
# [해석] 산도가 높은데 pH도 높은(불안정한) 샘플들이 존재함. 이를 통합한 '산도 안정성' 지표의 필요성.
```

### 1. 알코올과 밀도의 비율

```python
wine['alc_den_ratio'] = wine['alcohol'] / wine['density'] 
```

### 2. 알코올과 잔류 당분의 비율

```python
wine['fermentation_index'] = wine['alcohol'] / wine['residual sugar']
```

### 3. 전체 산도 중 휘발성 산도가 차지하는 비율

```python
wine['vol_acid_ratio'] = wine['volatile acidity'] / (wine['fixed acidity'] + wine['citric acid'])
```

### 4. 잔류당분과 밀도간의 관계 (밀도-당분 = 0 인 경우, quality가 낮음 _ 좋은와인X,설탕물O)

```python
wine['non_sugar_extract'] = wine['density'] - (wine['residual sugar'] * 0.0011)
```

### 5. 와인과 결합한 황만 남기고 다중공산성 제거 (+와인의 노후정도) - 추후 total sulfur 관련 col .drop

```python
wine['bound_sulfur'] = wine['total sulfur dioxide'] - wine['free sulfur dioxide']
```

### 6. 전체 황 중 Free sulfur의 비율 - 신선도 - 추후 total sulfur 관련 col .drop

```python
wine['free_sulfur_ratio'] = wine['free sulfur dioxide'] / wine['total sulfur dioxide'] 
```

### 7. 산도-알코올 밸런스

```python
wine['acid_alc_ratio'] = (wine['fixed acidity'] + wine['volatile acidity'] + wine['citric acid']) / wine['alcohol']
```

### 8. 산도와 알코올 간의 관계

```python
wine['acidity_stability'] = (wine['fixed acidity'] + wine['volatile acidity'] + wine['citric acid']) / wine['pH']
```

#### <b> 클래스 가중치(Class Weight) 커스텀 조정

```python
import numpy as np
import pandas as pd
from lightgbm import LGBMClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score, classification_report, confusion_matrix

def wine_score(df):
    # 1. [객관적 검증] 데이터 분할 (8:2) [cite: 2026-02-03]
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['quality'])
    
    # 2. [피처 정의] 각 타입별 최적 세트 [cite: 2026-02-04]
    red_feats = ['alcohol', 'pH', 'sulphates', 'log_residual sugar', 'log_chlorides', 
                 'log_total sulfur dioxide', 'vol_acid_ratio', 'acidity_stability', 
                 'non_sugar_extract', 'alc_den_ratio', 'free_sulfur_ratio', 'bound_sulfur', 'fermentation_index']
    white_feats = ['alcohol', 'pH', 'sulphates', 'log_volatile acidity', 'log_residual sugar', 
                   'log_chlorides', 'log_total sulfur dioxide', 'acidity_stability', 
                   'non_sugar_extract', 'alc_den_ratio', 'free_sulfur_ratio', 'bound_sulfur']







    
    # 3. [Stage 1] 와인 타입 분류 (Red/White) 학습
    common_feats = list(set(red_feats + white_feats))
    model_s1 = LGBMClassifier(n_estimators=100, random_state=42, verbose=-1)
    model_s1.fit(train_df[common_feats], train_df['type'])






    
    
    # 4. [Stage 2] 타입별 품질 예측 모델 학습 (커스텀 가중치 적용) [cite: 2026-02-04]
    
    # ------------------------ Red Wine Model ------------------------
    df_r = train_df[train_df['type'] == 0]
    le_r = LabelEncoder()
    y_r = le_r.fit_transform(df_r['quality'])
    
    # 오분류가 많은 4, 7, 8등급 벌점 강화 [cite: 2026-02-04]
    # 인덱스 순서: 3(0), 4(1), 5(2), 6(3), 7(4), 8(5)
    red_weights = {0: 2.0
                   , 1: 1.5
                   , 2: 1.0
                   , 3: 1.0
                   , 4: 1.3
                   , 5: 1.8}
    
    X_r_res, y_r_res = SMOTE(random_state=42, k_neighbors=2).fit_resample(df_r[red_feats], y_r)
    model_r = LGBMClassifier(n_estimators=521, learning_rate=0.0314, num_leaves=119, 
                             class_weight=red_weights, random_state=42, verbose=-1)
    model_r.fit(X_r_res, y_r_res)
    
    # ------------------------ White Wine Model ------------------------
    df_w = train_df[train_df['type'] == 1]
    le_w = LabelEncoder()
    y_w = le_w.fit_transform(df_w['quality'])
    
    # 5, 6등급 혼선 방지를 위한 하향 조정 및 양끝단 강화 [cite: 2026-02-04]
    # 인덱스 순서: 3(0), 4(1), 5(2), 6(3), 7(4), 8(5), 9(6)
    white_weights = {0: 2.0
                     , 1: 1.6
                     , 2: 0.9
                     , 3: 0.9
                     , 4: 1.4
                     , 5: 1.8
                     , 6: 2.5}
    
    X_w_res, y_w_res = SMOTE(random_state=42, k_neighbors=2).fit_resample(df_w[white_feats], y_w)
    model_w = LGBMClassifier(n_estimators=700, learning_rate=0.05, num_leaves=63, 
                             class_weight=white_weights, random_state=42, verbose=-1)
    model_w.fit(X_w_res, y_w_res)
    
    # 5. [최종 예측] 테스트 데이터 활용
    final_preds = []
    for _, row in test_df.iterrows():
        # 타입 예측 후 해당 모델 적용
        pred_type = model_s1.predict(row[common_feats].values.reshape(1, -1))[0]
        
        if pred_type == 0:
            p = model_r.predict(row[red_feats].values.reshape(1, -1))
            final_preds.append(le_r.inverse_transform(p)[0])
        else:
            p = model_w.predict(row[white_feats].values.reshape(1, -1))
            final_preds.append(le_w.inverse_transform(p)[0])
            
    # 6. [결과 출력] 통합 성능 리포트
    actual = test_df['quality'].values
    print(f"🚀 최종 통합 F1-Micro Score: {f1_score(actual, final_preds, average='micro'):.4f}")
    print("\n[Classification Report]")
    print(classification_report(actual, final_preds, zero_division=0))
    
    print("\n[Confusion Matrix]")
    labels = sorted(df['quality'].unique())
    cm = confusion_matrix(actual, final_preds, labels=labels)
    print(pd.DataFrame(cm, index=labels, columns=labels))

# ---------------------------------------------------------
# 실행 순서 가이드
# 1. EDA 직후 이상치 경계값 대치(Clipping) 실행
# 2. 로그 변환 및 파생 피처 생성 완료
# 3. wine_score(wine) 호출
# ---------------------------------------------------------
wine_score(wine)
```

## <b>검정 - 5-Fold 결과 분석

```python
from sklearn.model_selection import StratifiedKFold
import numpy as np

def wine_cv_score_original(df):
    # 1. 5-Fold 설정 데이터 5개 나눠서 5번검증
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    total_f1_scores = []
    
    # 피처 정의 영향주는 성분이 달라서 레드랑 화이트 다르게 지정
    red_feats = ['alcohol', 'pH', 'sulphates', 'log_residual sugar', 'log_chlorides', 
                 'log_total sulfur dioxide', 'vol_acid_ratio', 'acidity_stability', 
                 'non_sugar_extract', 'alc_den_ratio', 'free_sulfur_ratio', 'bound_sulfur', 'fermentation_index']
    white_feats = ['alcohol', 'pH', 'sulphates', 'log_volatile acidity', 'log_residual sugar', 
                   'log_chlorides', 'log_total sulfur dioxide', 'acidity_stability', 
                   'non_sugar_extract', 'alc_den_ratio', 'free_sulfur_ratio', 'bound_sulfur']
    common_feats = list(set(red_feats + white_feats))

    # [가중치 원상복구] 0.6931을 기록했던 설정입니다
    original_red_weights = {0: 2.0, 1: 1.5, 2: 1.0, 3: 1.0, 4: 1.3, 5: 1.8}
    original_white_weights = {0: 2.0, 1: 1.6, 2: 0.9, 3: 0.9, 4: 1.4, 5: 1.8, 6: 2.5}

    print(f"🔄 [가중치 복구] 5-Fold 평균 점수 산출 중...")

    for i, (train_idx, test_idx) in enumerate(skf.split(df, df['quality'])):
        train_df = df.iloc[train_idx]
        test_df = df.iloc[test_idx]
        
        # Stage 1: 타입 분류
        model_s1 = LGBMClassifier(n_estimators=100, random_state=42, verbose=-1)
        model_s1.fit(train_df[common_feats], train_df['type'])
        
        # Stage 2: 타입별 모델 (가중치 원복 적용)
        # RED
        df_r = train_df[train_df['type'] == 0]
        le_r = LabelEncoder()
        y_r = le_r.fit_transform(df_r['quality'])
        X_r_res, y_r_res = SMOTE(random_state=42, k_neighbors=2).fit_resample(df_r[red_feats], y_r)
        model_r = LGBMClassifier(n_estimators=521, learning_rate=0.0314, num_leaves=119, 
                                 class_weight=original_red_weights, random_state=42, verbose=-1)
        model_r.fit(X_r_res, y_r_res)
        
        # WHITE
        df_w = train_df[train_df['type'] == 1]
        le_w = LabelEncoder()
        y_w = le_w.fit_transform(df_w['quality'])
        X_w_res, y_w_res = SMOTE(random_state=42, k_neighbors=2).fit_resample(df_w[white_feats], y_w)
        model_w = LGBMClassifier(n_estimators=700, learning_rate=0.05, num_leaves=63, 
                                 class_weight=original_white_weights, random_state=42, verbose=-1)
        model_w.fit(X_w_res, y_w_res)
        
        # 예측
        fold_preds = []
        for _, row in test_df.iterrows():
            pred_type = model_s1.predict(row[common_feats].values.reshape(1, -1))[0]
            if pred_type == 0:
                p = model_r.predict(row[red_feats].values.reshape(1, -1))
                fold_preds.append(le_r.inverse_transform(p)[0])
            else:
                p = model_w.predict(row[white_feats].values.reshape(1, -1))
                fold_preds.append(le_w.inverse_transform(p)[0])
        
        f1 = f1_score(test_df['quality'], fold_preds, average='micro')
        total_f1_scores.append(f1)
        print(f"📍 Fold {i+1} 점수: {f1:.4f}")

    print(f"\n✨ [최종 성적] 5-Fold 평균 F1-Micro: {np.mean(total_f1_scores):.4f}")
    
# 실행
wine_cv_score_original(wine)
```
