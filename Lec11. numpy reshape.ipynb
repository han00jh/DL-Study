# <b>1D : list vs arrary vs series
* [1,2,3,4] 리스트는 shape없다
* [1 2 3 4] 콤마차이 shape은 arrary꺼

```python
import numpy as np
import pandas as pd
```

```python
num_list = [1,2,3,4]
print(  len(num_list)  , np.ndim(num_list) , num_list) #  np.ndim(num_list): 차수가 몇인지
```

```python
num_list = [1,2,3,4]
num_arr = np.array(num_list) # list를 arr로 바꾸기
print(  len(num_arr)  , np.ndim(num_arr) , num_arr)
print( num_arr.shape) # num_arr = numpy.ndarray
```

```python
type( (4) ), type( (4,) ) # 두 개 타입 다름
```

```python
s = pd.Series(num_list)
s = pd.Series(num_arr)
s # 1차 디멘션이면 시리즈 만들기 가능
```

# <b> reshape() vs resize()
* reshape() : 원본은 두고 변환한 행열을 반환
* resize() : 원본을 변환, 리턴없음

```python
arr = np.arange(12)
a= arr.reshape(2,6) #원본 안건들고 복사해서 사용
arr, a
```

```python
arr = np.arange(12)
print(arr)
a= arr.resize(2,6) #resize는 return없음. 원본 그대로 바꾼다 그래서 reshape 주로 사용
print(arr)
```

# <b>차수 확장(expand)
* <font color = red><b> -1
* np.newaxis - 슬라이싱처럼 차원 추가
* numpy.expand_dims(a,axis) - 차원 추가 전용

```python
arr = np.arange(12)  # 1차
arr
```

```python
#a = arr.reshape(2,-1) # 1차 -> 2차
a = arr.reshape(-1,1) # 1차 -> 2차
a
```

```python
a = arr[:, np.newaxis]
a
```

```python
print(arr, arr.shape)
a = np.expand_dims(arr, axis = 0) #행
print(a, a.shape)
```

```python
print(arr, arr.shape)
a = np.expand_dims(arr, axis = 1) #열
print(a, a.shape)
```

```python
print(arr, arr.shape)
a = np.expand_dims(arr, axis = (0,1)) #행도 추가 열도 추가
print(a, a.shape)
```

# <font color =red><b> 차수 축소(expand)
* <b><font color =red><b> -1
* <font color =red><b> 1차 : flatten() == ravel()
* np.squeeze(arr, axis) =  shape에서 1인 차원 삭제

```python
arr = np.arange(6).reshape(1,2,3) #3차
arr
```

```python
a = arr.reshape(-1)  # 1차
# == arr.flatten() 1차
# == arr.ravel() 1차
a
```

```python
a = arr.reshape(-1,1)  # 3차 -> 2차
a
```

```python
print(arr, arr.shape)
a = np.squeeze(arr, axis = 0) #3차 ->2차
a = np.squeeze(arr)
print(a, a.shape)
```

# <font color=red><b>합치기(stack </b></font> == concatenate())
* <b><font color=red>hstack()
* <b><font color=red>vstack()
* <b><font color=red>column_stack()


https://rfriend.tistory.com/352

```python
a = [1,2,3]
b = [10,20,30]
```

```python
res = np.hstack(  [a,b]  )
# == np.flatten(  [a,b]  )
res
```

```python
res = np.column_stack(  [a,b]  )
#== np.stack(  [a,b]  , axis = 1)
res
```
