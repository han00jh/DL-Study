# 🐳 Docker 개념 간단 정리

## Docker란?
> 애플리케이션을 **컨테이너**라는 격리된 환경에 패키징해서 어디서든 동일하게 실행할 수 있게 해주는 플랫폼

"내 컴퓨터에서는 되는데..." 문제를 해결해 준다.

---

## 핵심 개념

### 📦 Image (이미지)
- 컨테이너를 만들기 위한 **읽기 전용 템플릿**
- 레이어(layer) 구조로 구성
- 예: `ubuntu:22.04`, `node:18`, `python:3.11`

### 🏃 Container (컨테이너)
- 이미지를 실행한 **인스턴스**
- 격리된 프로세스 환경
- 이미지 1개 → 컨테이너 N개 생성 가능

### 📄 Dockerfile
- 이미지를 만들기 위한 **설계도 (스크립트)**
- 어떤 OS, 어떤 패키지, 어떤 명령을 실행할지 정의

```dockerfile
FROM node:18
WORKDIR /app
COPY . .
RUN npm install
CMD ["node", "index.js"]
```

### 🗂️ Docker Hub
- 이미지를 저장하고 공유하는 **원격 저장소**
- GitHub의 이미지 버전이라고 보면 됨
- [hub.docker.com](https://hub.docker.com)

### + "이미지"라고 부르는 이유
 - 이미지는 실행 전 상태이기 때문.
 - 이미지 = 설치된 OS, 파일, 설정이 담긴 스냅샷 (정지된 상태)
 - 컨테이너 = 이미지를 실행한 상태

     - 이 둘을 구분하는 게 Docker의 핵심 개념.

<b>비유하면:




틀(이미지) 하나로 붕어빵(컨테이너)을 100개 찍어낼 수 있다.



### 🔗 Docker Compose
- **여러 컨테이너**를 한 번에 정의하고 실행
- `docker-compose.yml` 파일로 관리
- 예: 앱 서버 + DB + Redis를 함께 구동

```yaml
services:
  app:
    build: .
    ports:
      - "3000:3000"
  db:
    image: postgres:15
    environment:
      POSTGRES_PASSWORD: secret
```

## VM vs Docker

| 항목 | VM (가상머신) | Docker (컨테이너) |
|------|-------------|-----------------|
| 크기 | GB 단위 | MB 단위 |
| 시작 속도 | 분 단위 | 초 단위 |
| OS | 각각 독립 OS | 호스트 OS 공유 |
| 격리 수준 | 완전 격리 | 프로세스 격리 |
| 성능 | 상대적으로 낮음 | 거의 네이티브 수준 |

## 자주 쓰는 명령어

```bash
# 이미지 다운로드
docker pull nginx

# 컨테이너 실행
docker run -d -p 8080:80 nginx

# 실행 중인 컨테이너 목록
docker ps

# 컨테이너 중지 / 삭제
docker stop <id>
docker rm <id>

# 이미지 목록 / 삭제
docker images
docker rmi <image>

# 이미지 빌드
docker build -t my-app .

# 컨테이너 내부 접속
docker exec -it <id> bash

# Compose 실행 / 종료
docker compose up -d
docker compose down
```

### 문장 자르기

#### <font color=red size=5><b>여러문장 : Tokenizer ★★★★★★★★★★

````python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

text = ["나는 오늘 친구랑 학교에 간다." 
        ,"너는 오늘 집에 간다"]

tk = Tokenizer()
tk.fit_on_texts(text)                                #-------------------[1]문장/단어자르기
#문장 자르면 단어사전구축가능
print(tk.word_index)                                 #-----------사전확인 : 단어 인덱스
print(tk.word_counts)                                 #-----------사전확인 : 단어 빈도

res = tk.texts_to_sequences(text)                    #-------------------[2]임베딩(단어자르기->인코딩->리스트)
print(res)                                           #-----------임베딩(인코딩된단어집합)확인 


pad = pad_sequences(res, maxlen=5,  padding='post')  #--------------------[3]패딩
print(pad)
````

* <font color=blue><b>TextVectorization

````python
from tensorflow.keras.layers import TextVectorization

text = ["나는 오늘 친구랑 학교에 간다." 
        ,"너는 오늘 집에 간다"]

vec = TextVectorization(output_mode="int" , output_sequence_length=5)
vec.adapt(text)                        #------------------ fit :: [1][2][3]문장자르기,단어자르기,패딩

res = vec(text)                        #------------------ transform :: 
print(res)

vocab = vec.get_vocabulary()           #-----------사전확인
#print(vocab)
word_index = {word: idx for idx, word in enumerate(vocab)}
print(word_index)
````

# <b>[2]  어간 추출(stemming) , 표제어 추출(lemmatizaiton)

* 정제(cleaning)
    - 특수기호 , <font color=red><b>불용어</b></font> (a is the 이 그 저 이다....) -> 노이즈 제거<br><br>
* 정규화(normalization) 
    - 표현이 다른 단어들을 한 단어로 표현 (빨갛고 붉게 붉은 --> 빨갛다)
    - <font color=red><b>어간(Stemming), 표제어(Lemmatization) 추출</b></font> --->  단어수 줄이기

| 구분 | 어간 추출 (Stemming) | 표제어 추출 (Lemmatization) |
| :--- | :--- | :--- |
| **핵심 개념** | 정해진 규칙에 따라 단어 뒷부분을 자름 | 단어의 형태학적 분석을 통해 원형을 찾음 |
| **특징** | 속도가 매우 빠름 (단순 컷팅) | 속도가 상대적으로 느림 (문맥 고려) |
| **결과물** | 사전에 없는 단어일 수 있음 (예: `formali`) | 항상 사전에 있는 표준 단어 (예: `formal`) |
| **정확도** | 낮음 (대충 자름) | 높음 (사전적 의미 유지) |
| **도구** | `PorterStemmer`, `LancasterStemmer` | `WordNetLemmatizer` |

## <b>nltk

### 어간 추출(stemming)
* formalize → formal
* allowance → allow
* electricical → electric

````python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()

tokenized_sentence = word_tokenize('Bones')
print('어간 추출 전 :', tokenized_sentence)
print('어간 추출 후 :',stemmer.stem('Bones'))
````

### 문장 내 불용어 제거

````python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english')) 

word_list = word_tokenize(example)

result = []
for word in word_list: 
    if word not in stop_words: 
        result.append(word) 

print('불용어 제거 전 :',word_list) 
print('불용어 제거 후 :',result)
````
