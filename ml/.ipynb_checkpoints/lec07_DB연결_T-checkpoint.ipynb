# <b>SQLAlchemy ★

## DB 연결

* <font color=blue><b>SQLAlchemy + Oracle 연결

```python
import pandas as pd
from sqlalchemy import create_engine    , text, Integer, Float, Numeric
from sqlalchemy.dialects.oracle import NUMBER,  FLOAT as ORA_FLOAT
```

```python
engine = create_engine("oracle+cx_oracle://it:0000@localhost:1521/xe")
##conn = engine.connect()
```

## C : 입력

```python
with engine.begin() as conn:
    conn.execute(text("INSERT INTO emp (empno, ename) VALUES (:aa, :bb)"),
                 {"aa": 9002, "bb": "KIM"}
                )
```

```python
with engine.begin() as conn:
    conn.execute(text("INSERT INTO emp (empno, ename, sal) VALUES (:aa, :bb, :cc)"),
                 { "aa": 9003
                  , "bb": "KIM"
                  , "cc" : 7.7
                 }
                )
```

## R : 조회

```python
with engine.connect() as conn:
    rows = conn.execute( text("SELECT empno, ename FROM emp") )
    for r in rows:
        print(r.empno, r.ename)
```

```python
sql = """
INSERT INTO emp (empno, ename, sal)
VALUES (:empno, :ename, :sal)
"""

with engine.begin() as conn:
    conn.execute(text(sql), {
        "empno": 9001,
        "ename": "KIM",
        "sal": 3000
    })
```

## R : where 조회

```python
with engine.connect() as conn:
    rows = conn.execute( text("SELECT empno, ename FROM emp where deptno = :aa and job=:bb"),
                                                                          {"aa": 20, "bb":'CLERK'} 
                       )
    for r in rows:
        print(r.empno, r.ename)
```

## DB 연결

```python
import cx_Oracle

conn = cx_Oracle.connect(
    user="it",
    password="0000",
    dsn="localhost:1521/xe"
)
cur = conn.cursor()
```

## C : 입력

```python
sql = """
    INSERT INTO emp (empno, ename)
    VALUES (:1, :2)
"""
cur.execute( sql  , (8899, "KIM"))
conn.commit()
```

## R : 조회

```python
cur.execute("SELECT empno, ename FROM emp")
rows = cur.fetchall()
for r in rows:
    print(r)
```

## U : 수정

```python
cur.execute("""
    UPDATE emp
    SET ename   = :1
    WHERE empno = :2
""", ("LEE", 8899))
conn.commit()
```

## D : 삭제

```python
cur.execute("""
    DELETE FROM emp
    WHERE empno = :1
""", (8899,))
conn.commit()
```

## DB 연결 닫기

```python
cur.close()
conn.close()
```
