## DB연결

```python
import pandas as pd
from sqlalchemy import create_engine,            text, Integer, Float, Numeric, String
from sqlalchemy.dialects.oracle import NUMBER,  FLOAT as ORA_FLOAT
```

```python
engine = create_engine("oracle+cx_oracle://it:0000@localhost:1521/xe")
conn = engine.connect()
```

## R : 조회 df.read_sql()

```python
df = pd.read_sql("SELECT * FROM emp", conn)
#print(df["EMPNO"])
df.head(2)
```

## C : 입력 df.to_sql()
* 테이블이 없을경우 생성해서 넣기 
* 테이블이 있을경우에러 -> 기존테이블 삭제 또는 다른 테이블로 변경

```python
df[['empno','ename','job']].to_sql(
    name="emp2",
    con=engine,
    if_exists="append",   # append | replace | fail
    index=False
) #오라클은 한번 넣으면 끝.싹 갈아엎는거 이외에는 불가
```

```python
df.to_sql(
    name="emp5",
    con=engine,
    if_exists="append",   # append | replace | fail
    index=False,

     dtype={
        'empno'   : Integer()
        ,"ename"  : String(20)
        ,"job"    : String(20)
        ,'sal'    : ORA_FLOAT(binary_precision=10)  #----- 실수 넣기(비추)
        ,'mgr'    : Numeric(10, 2)                  #----- 실수 넣기
        ,'comm'   : NUMBER(10, 2)                   #----- 실수 넣기
        ,'deptno' : NUMBER(10, 2)                   #----- 실수 넣기
    }
)
```

## DB 연결 닫기
* 반드시 닫아야함(순서주의)

```python
conn.close()
engine.dispose()
```

# sqlAIchemy★

## DB연결

```python
import pandas as pd
from sqlalchemy import create_engine,            text, Integer, Float, Numeric, String
from sqlalchemy.dialects.oracle import NUMBER,  FLOAT as ORA_FLOAT
```

```python
engine = create_engine("oracle+cx_oracle://it:0000@localhost:1521/xe")
#conn = engine.connect()
```

## C : 입력

```python
with engine.begin() as conn:
    conn.execute(text("INSERT INTO emp (empno, ename) VALUES (:aa, :bb)"),
                 {"aa": 9002, "bb": "KIM"}
                )
```

```python
with engine.begin() as conn:
    conn.execute(text("INSERT INTO emp (empno, ename, sal) VALUES (:aa, :bb, :cc)"),
                 { "aa": 9003
                  , "bb": "KIM"
                  , "cc" : 7.7
                 }
                )
```

## R : 조회

```python
with engine.connect() as conn:
    rows = conn.execute(
        text("SELECT empno, ename FROM emp")
    )
    for r in rows:
        print(r.empno, r.ename) #with절 사용으로 자동 close개꿀
```

```python
sql = """
INSERT INTO emp (empno, ename, sal)
VALUES (:empno, :ename, :sal)
"""

with engine.begin() as conn:
    conn.execute(text(sql), {
        "empno": 9001,
        "ename": "KIM",
        "sal": 3000
    })
```

## R : where 조건절 조회

```python
with engine.connect() as conn:
    rows = conn.execute( text("SELECT empno, ename FROM emp where deptno = :aa and job=:bb"),
                                                                          {"aa": 20, "bb":'CLERK'} 
                       )    # 다건조회시 .fetchall()

    for r in rows :
        print(r.empno, r.ename)
```

## DB연결

```python
import cx_Oracle

conn = cx_Oracle.connect(
    user="it",
    password="0000",
    dsn="localhost:1521/xe"
)
cur = conn.cursor()
```

## C : 입력

```python
sql = """
    INSERT INTO emp (empno, ename)
    VALUES (:1, :2)
"""
cur.execute(sql , (8899, "KIM"))

conn.commit()
```

## R : 조회

```python
cur.execute("SELECT empno, ename FROM emp")
rows = cur.fetchall()
for r in rows:
    print(r)
```

## U : 수정

```python
cur.execute("""
    UPDATE emp
    SET ename = :1
    WHERE empno = :2
""", ("LEE", 8899))

conn.commit()
```

## D : 삭제

```python
cur.execute("""
    DELETE FROM emp
    WHERE empno = :1
""", (8899,))

conn.commit()
```

## DB 연결 닫기
* 반드시 닫아야함(순서주의)

```python
cur.close()
conn.close()
```
