# <b> REST(Representational State Transfer)
* <b>분산시스템 아키텍쳐(구조)
-------------------------------------------
<pre>
* 요청 어떻게?
  -제공자(provider)가 정한 요청URL, 규격, KEY를 가지고 접속해야 데이터가 반환됨
* 요청 아무거나 다 가능?
    - 무조건 웹서비스만 가능
        * 웹 : 세션(session)/쿠키(cookie)를 이용한 사용자 인증/인가
        * REST : KEY를 이용한 사용자 인증/인가
* 요청 데이터 형태?
  -리턴하는 데이터는 일반적으로 JSON
        - JSON(JavsScript Object Notaion) : 객체 표현 문법
        - {"age" : 10} dic형식
* 요청 통신 방식?
  -HTTP(S) 프로토콜의 GET/POST 방식의 통신만 지원
  -동기(sync)키오스크/비동기(async)진동벨 가능하나 일반적으로 <b>비동식 통신</b>

# <b> REST API

* 클라이언트와 서버간 REST 서비스를 제공하기 위한 API(기능, 함수)

# <b> RESTful

* 아키텍쳐의 <b>설계 원칙과 제약 조건을 준수</b>하여
* 클라이언트와 서버간의 데이터를 <b>직관적이고 표준화된 방식</b>으로 주고받는
* REST 서비스를 제공하기 위한 API(기능,함수)

# <b> Web GET vs. POST 통신방식

| 구분           | GET 방식                   | POST 방식                           |
| ------------ | ------------------------ | --------------------------------- |
| 데이터 전송 위치    | **HTTP Header** <br> URL 뒤 **Query String(?a=10&b=kim)**   | **HTTP Body**                     |
| 예시           | `/search?q=apple&page=1` | Body에 JSON / Form 데이터             |
| 데이터 노출       | **URL에 그대로 노출됨**         | URL에 노출 ❌                         |
| 보안성          | 낮음 (로그, 히스토리 노출)         | 상대적으로 높음                          |
| 데이터 길이 제한    | 있음 (브라우저·서버 제한)          | 사실상 제한 없음                         |
| 사용 예         | 검색, 페이지 조회<br>브라우저에서 직접 주소적기, &lt;a href=''&gt;간단링크             | 로그인, 회원가입, 저장 <br> <b>&lt;Form&gt;태그를 활용한 버튼 post 방식</b>   |
